```python
xx = np.linspace(1, 10, 100)
yy = np.sqrt(xx)
xi = np.arange(12)[1:11]
yi = np.sqrt(xi)
fig, ax = plt.subplots(1, 1)
ax.plot(xx, yy, 'b--', linewidth=1.0, label=r'$\sqrt(N)$')
ax.plot(xi, yi, 'bo', markersize=4.0)
ax.plot(xx, np.ones(12))
plt.show()
```
